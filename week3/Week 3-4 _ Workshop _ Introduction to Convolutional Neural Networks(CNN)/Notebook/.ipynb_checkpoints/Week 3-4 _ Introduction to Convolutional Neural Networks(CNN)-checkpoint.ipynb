### 2. Feature Extraction Methods for 2D Images
##### Histogram of Oriented Gradients (HOG):

- HOG looks at the changes in brightness (gradients) in an image.
- It creates a histogram (like a graph) that shows how often these gradients occur in different directions.
- HOG is useful for detecting objects, recognizing faces, and understanding the texture of an image.

##### Local Binary Patterns (LBP):

- LBP looks at the texture of an image by comparing each pixel to its neighboring pixels.
- It creates a pattern that describes the texture around each pixel.
- LBP is good at recognizing faces, classifying textures, and analyzing medical images, even when there are changes in lighting.

##### Scale-Invariant Feature Transform (SIFT):

- SIFT finds unique points in an image called keypoints.
- It creates a description of the area around each keypoint based on the gradients (changes in brightness).
- SIFT is useful for stitching images together, tracking objects, and creating panoramic images, even when the scale, rotation, or lighting changes.

##### Convolutional Neural Networks (CNNs):

- CNNs are a type of machine learning model that can learn to extract features from images on their own.
- They look at the raw pixel data and learn to recognize patterns and features that are useful for tasks like classifying images, detecting objects, and separating different parts of an image (segmentation).
- CNNs can automatically learn complex and abstract features that are difficult for humans to define manually.



### 3. Feature Extraction Methods for 3D Images

### Voxel-Based Methods:

- Analyze 3D volume data directly.
- Extract features from voxel intensities, gradients, or textures.
- Use techniques like voxelization, region growing, morphological operations.

### Shape-Based Methods:

- Focus on geometric properties of 3D objects/structures.
- Extract features like surface curvature, volume, compactness.
- Use mesh generation, surface fitting, geometric descriptors.

### Deep Learning for 3D Data:

- Process 3D volumes directly.
- Learn hierarchical features from 3D patches/volumes.
- Used for organ segmentation, tumor detection, disease classification.


# Understanding Filters and Convolution


Filters are mathematical operations applied to input data (e.g., images, signals) to enhance, suppress, or isolate certain characteristics or patterns of interest.

<center><img src="https://www.researchgate.net/publication/334058390/figure/fig1/AS:960484207513610@1606008688319/The-mathematical-explanation-of-the-convolution-operation.png" width="600" height="400"></center>

### Role of Filters in Feature Extraction

- **Noise Reduction**: Filters can help remove unwanted disturbances or interference from signals or images, making it easier to focus on the important information.

- **Feature Enhancement**: Filters can make specific parts of a signal or image stand out more clearly, making them easier to analyze.

- **Edge Detection**: Some filters can highlight the boundaries or edges between different areas in an image, making it simpler to identify distinct regions or objects.

- **Pattern Recognition**: Filters can be created to identify and highlight specific patterns or shapes within signals or images. For example, a special type of filter can be used to find a particular pattern within an image.

source (https://poloclub.github.io/cnn-explainer/)

- **Filter Size**:
  - **Small Filters**: Filters that are small in size, like 3x3 or 5x5, are good for capturing and detecting small details and patterns in an image.
  - **Large Filters**: Filters that are larger in size, like 7x7 or 9x9, are better at capturing and detecting bigger patterns and structures across the whole image.
  - **Computation**: Larger filters require more computing power because they process a bigger area of the image. Smaller filters are faster but may miss some of the bigger patterns.

# Convolutional Neural Networks (CNNs)


### CNN Architecture

1. **Input Layer**: Receives the raw image data.

2. **Convolutional Layers**: Use filters to extract features from the image.

3. **Activation Function**: Introduces non-linearity to the network.

4. **Pooling Layer**: Reduces spatial dimensions while retaining important information.

5. **Fully Connected Layers**: Learn complex patterns from extracted features.

6. **Output Layer**: Provides final predictions, often using softmax for classification tasks.

 Typical structure of a convolutional neural network:

`Input layer -> [Convolutional layer -> activation layer -> pooling layer] -> Output layer`

# Transpose Convolution and Upsampling

Transposed convolution, also known as deconvolution or fractionally strided convolution, is a technique commonly used for upsampling or generating high-resolution outputs from low-resolution inputs in deep learning models, particularly in tasks like image super-resolution, image segmentation, and generative modeling.

Imagine you have a small image, like a 4x4 pixel image. You want to make this image bigger, like an 8x8 pixel image. One way to do this is to use a technique called transpose convolution.

Here's how it works:

1. First, you take the small 4x4 image and add some extra space around it by padding it with zeros. This creates a slightly larger 6x6 image.
2. Next, you have a special filter, like a 3x3 square. This filter is going to be used to "convolve" (or slide around) the padded 6x6 image.
3. But before you convolve the filter, you flip it upside down and backwards, so it's like a mirror image of the original filter.
4. Then you slide this flipped filter around the padded 6x6 image, and at each position, you multiply the filter values with the image values and add them up. This gives you one pixel in the final 8x8 output image.
5. You repeat this process, sliding the flipped filter around the entire padded 6x6 image, to generate all the pixels in the final 8x8 output image.

The key idea is that by flipping the filter and sliding it around, you can take a small image and make it bigger, in a way that the network can learn to optimize. This is called "transpose convolution" because it's the reverse of the normal convolution operation used for making images smaller.


```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Define the model
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        # Define the convolutional layer
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=3, stride=1, padding=1)
        # Define the transpose convolutional layer
        self.deconv1 = nn.ConvTranspose2d(in_channels=1, out_channels=1, kernel_size=3, stride=2, padding=1, output_padding=1)

    def forward(self, x):
        # Forward pass through the convolutional layer
        x = self.conv1(x)
        # Forward pass through the transpose convolutional layer
        x = self.deconv1(x)
        return x

# Create an instance of the model
model = Model()

# Assuming train_loader is defined somewhere
# Get a batch of images and labels
images, labels = next(iter(train_loader))

# Apply the convolutional layer to the images
conv_images = model.conv1(images)

# Apply the transpose convolutional layer to the convolved images
deconv_images = model.deconv1(conv_images)

# Detach the tensors for plotting
conv_images = conv_images.detach()
deconv_images = deconv_images.detach()
images = images.detach()

# Plot the original, convolved, and de-convolved images
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(images[0].squeeze(), cmap='gray')
plt.title('Original Image')
plt.subplot(1, 3, 2)
plt.imshow(conv_images[0].squeeze(), cmap='gray')
plt.title('Convolved Image')
plt.subplot(1, 3, 3)
plt.imshow(deconv_images[0].squeeze(), cmap='gray')
plt.title('De-convolved Image')
plt.show()

# Print shapes of tensors
print(images.shape)
print(conv_images.shape)
print(deconv_images.shape)
```

# Activation Function

In the context of Convolutional Neural Networks (CNNs), an activation function is a mathematical function applied to the output of each neuron in the network. The activation function introduces non-linearity into the neural network, allowing it to learn and model complex, non-linear relationships in the input data.

The primary purpose of an activation function in a CNN is to introduce non-linearity, which is essential for the network to learn and approximate complex functions. Without an activation function, the output of a CNN would simply be a linear combination of the inputs, limiting its ability to model non-linear patterns in the data.

# Pooling Layers


Convolutional layers in CNNs break down an input (like an image) into smaller pieces and identify different features, represented as feature maps. However, having too many feature maps can make processing difficult.

Pooling layers help simplify and summarize the information in these feature maps. They downsample the feature maps by:

- **Max pooling**: Keeping only the maximum value from small groups of values in each feature map.
- **Average pooling**: Calculating the average of small groups of values and using that average to represent each group.

```python
class TinyVGG(nn.Module):
    """
    Model architecture copying TinyVGG from:
    https://poloclub.github.io/cnn-explainer/
    """
    def __init__(self, input_shape: int, hidden_units: int, output_shape: int) -> None:
        super().__init__()
        self.conv_block_1 = nn.Sequential(
            nn.Conv2d(in_channels=input_shape,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1, # default
                      padding=1), # options = "valid" (no padding) or "same" (output has same shape as input) or int for specific number
            nn.ReLU(),
            nn.Conv2d(in_channels=hidden_units,
                      out_channels=hidden_units,
                      kernel_size=3,
                      stride=1,
                      padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2,
                         stride=2) # default stride value is same as kernel_size
        )
        self.conv_block_2 = nn.Sequential(
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(hidden_units, hidden_units, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=hidden_units*31*31,
                      out_features=output_shape)
        )

    def forward(self, x: torch.Tensor):
      return self.classifier(self.conv_block_2(self.conv_block_1(x)))

torch.manual_seed(42)
model = TinyVGG(input_shape=3, # number of color channels (3 for RGB)
                  hidden_units=10,
                  output_shape=len(train_data.classes)).to(device)
model
```

```python
def train_step(model: torch.nn.Module,
               dataloader: torch.utils.data.DataLoader,
               loss_fn: torch.nn.Module,
               optimizer: torch.optim.Optimizer):
    # Put model in train mode
    model.train()

    # Setup train loss and train accuracy values
    train_loss, train_acc = 0, 0

    # Loop through data loader data batches
    for batch, (X, y) in enumerate(dataloader):
        # Send data to target device
        X, y = X.to(device), y.to(device)

        # 1. Forward pass
        y_pred = model(X)

        # 2. Calculate  and accumulate loss
        loss = loss_fn(y_pred, y)
        train_loss += loss.item()

        # 3. Optimizer zero grad
        optimizer.zero_grad()

        # 4. Loss backward
        loss.backward()

        # 5. Optimizer step
        optimizer.step()

        # Calculate and accumulate accuracy metric across all batches
        y_pred_class = torch.argmax(torch.softmax(y_pred, dim=1), dim=1)
        train_acc += (y_pred_class == y).sum().item()/len(y_pred)

    # Adjust metrics to get average loss and accuracy per batch
    train_loss = train_loss / len(dataloader)
    train_acc = train_acc / len(dataloader)
    return train_loss, train_acc
```

```python
def test_step(model: torch.nn.Module,
              dataloader: torch.utils.data.DataLoader,
              loss_fn: torch.nn.Module):
    # Put model in eval mode
    model.eval()

    # Setup test loss and test accuracy values
    test_loss, test_acc = 0, 0

    # Turn on inference context manager
    with torch.inference_mode():
        # Loop through DataLoader batches
        for batch, (X, y) in enumerate(dataloader):
            # Send data to target device
            X, y = X.to(device), y.to(device)

            # 1. Forward pass
            test_pred_logits = model(X)

            # 2. Calculate and accumulate loss
            loss = loss_fn(test_pred_logits, y)
            test_loss += loss.item()

            # Calculate and accumulate accuracy
            test_pred_labels = test_pred_logits.argmax(dim=1)
            test_acc += ((test_pred_labels == y).sum().item()/len(test_pred_labels))

    # Adjust metrics to get average loss and accuracy per batch
    test_loss = test_loss / len(dataloader)
    test_acc = test_acc / len(dataloader)
    return test_loss, test_acc
```

```python
from tqdm.auto import tqdm

# 1. Take in various parameters required for training and test steps
def train(model: torch.nn.Module,
          train_dataloader: torch.utils.data.DataLoader,
          test_dataloader: torch.utils.data.DataLoader,
          optimizer: torch.optim.Optimizer,
          loss_fn: torch.nn.Module = nn.CrossEntropyLoss(),
          epochs: int = 5):

    # 2. Create empty results dictionary
    results = {"train_loss": [],
        "train_acc": [],
        "test_loss": [],
        "test_acc": []
    }

    # 3. Loop through training and testing steps for a number of epochs
    for epoch in tqdm(range(epochs)):
        train_loss, train_acc = train_step(model=model,
                                           dataloader=train_dataloader,
                                           loss_fn=loss_fn,
                                           optimizer=optimizer)
        test_loss, test_acc = test_step(model=model,
            dataloader=test_dataloader,
            loss_fn=loss_fn)

        # 4. Print out what's happening
        print(
            f"Epoch: {epoch+1} | "
            f"train_loss: {train_loss:.4f} | "
            f"train_acc: {train_acc:.4f} | "
            f"test_loss: {test_loss:.4f} | "
            f"test_acc: {test_acc:.4f}"
        )

        # 5. Update results dictionary
        results["train_loss"].append(train_loss)
        results["train_acc"].append(train_acc)
        results["test_loss"].append(test_loss)
        results["test_acc"].append(test_acc)

    # 6. Return the filled results at the end of the epochs
    return results
```

```python
def make_predictions(model: torch.nn.Module, data: list, device: torch.device = device):
    pred_probs = []
    model.eval()
    with torch.inference_mode():
        for sample in data:
            # Prepare sample
            sample = torch.unsqueeze(sample, dim=0).to(device) # Add an extra dimension and send sample to device

            # Forward pass (model outputs raw logit)
            pred_logit = model(sample)

            # Get prediction probability (logit -> prediction probability)
            pred_prob = torch.softmax(pred_logit.squeeze(), dim=0)

            # Get pred_prob off GPU for further calculations
            pred_probs.append(pred_prob.cpu())

    # Stack the pred_probs to turn list into a tensor
    return torch.stack(pred_probs)
```

**Summary of Key Concepts Covered:**

- **Convolutional Neural Networks (CNNs):** These are specialized neural networks designed for processing grid-like data, such as images. They mimic the receptive fields found in the human visual cortex, detecting simple patterns first and progressively more complex patterns.
  
- **CNN Architecture:** Typically consists of three main layers:
  - **Convolutional Layer:** Performs a dot product operation between a kernel and the input data, extracting features.
  - **Pooling Layer:** Reduces the spatial size of the representation, decreasing computational load, and providing some translation invariance.
  - **Fully Connected Layer:** Neurons in this layer have full connectivity with all neurons in the preceding and succeeding layers.
  
- **Motivation behind Convolution:**
  - **Sparse Interaction:** CNNs have sparse interaction, reducing the memory requirement and improving statistical efficiency.
  - **Parameter Sharing:** Neurons in CNNs share parameters, allowing them to use the same set of weights for different spatial points.
  - **Equivariant Representation:** Layers in CNNs have a property of equivariance to translation, meaning that if the input changes, the output changes in the same way.
  
- **Non-Linearity Layers:** Introduced after convolutional layers to introduce non-linearity into the activation maps, commonly using functions like Sigmoid, Tanh, or ReLU.
  
- **Designing CNNs:** Building a CNN involves defining the architecture, including convolutional layers, pooling layers, fully connected layers, and non-linearity layers, followed by training using techniques like batch normalization and optimization algorithms like Adam.
  
- **Applications of CNNs:** CNNs are widely used in various applications, including object detection, semantic segmentation, and image captioning, among others.

This summary covers the fundamental concepts of CNNs, their architecture, motivations behind their design, practical implementation, and applications in real-world scenarios.
